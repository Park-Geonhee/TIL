```python
song_data = []
rank = 1
songs = soup.select('tbody>tr')
for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    #print(title, singer, sep = "|")
    mylist=['melon',rank,title,singer]
    song_data.append(mylist)
    rank+=1
    
song_data
```

```python
bugs_data = []

for bugs in bugs_top:
    rank = int(bugs.select('strong')[0].text)
    title = bugs.select('p.title>a')[0].text
    singer = bugs.select('p.artist>a')[0].text
    print('{0}위'.format(rank),title,singer,sep=' | ')
    bugs_data.append(['bugs',rank,title,singer])
    
bugs_data
```

```python
bugss = pd.DataFrame(bugs_data, columns = ['서비스','순위','타이틀','가수'],
                    )
bugss.to_excel('bugs_class.xlsx', index=False)
```

## 지니뮤직 웹크롤링

```python
ser = Service('../chromedriver/chromedriver.exe')

driver = webdriver.Chrome(service = ser)

url = 'https://www.genie.co.kr/chart/top200'
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
soup.head()
```

```python
chart = soup.select('tbody>tr')
chart[:10]
```

```python
len(chart)
```

```python
chart[0].select('a.title.ellipsis')[0].text.strip()
```

```python
chart[0].select('a.artist.ellipsis')[0].text
```

```python
chart[0].select('td.number')[0].text.replace('\n','').replace(' ','')
# 랭크는 따로 for문 이용
```

```python
list = []
rank = 1
for genie in chart:
    title = genie.select('a.title.ellipsis')[0].text.strip()
    singer = genie.select('a.artist.ellipsis')[0].text
    print('{0}위'.format(rank),title,singer, sep = " | ")
    list.append(['genie',rank,title,singer])
    rank +=1

list
```

```python
genieDF = pd.DataFrame(list, columns = ['서비스','순위','타이틀','가수'])
genieDF.to_excel('genie_class.xlsx',index=False)
```
