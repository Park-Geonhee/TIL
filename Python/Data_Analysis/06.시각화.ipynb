```python
def create_kto_data(yy,mm):
    file_path = './files/kto_{}{}.xlsx'.format(yy, mm)
    df = pd.read_excel(file_path, header = 1, skipfooter = 4,
                      usecols = 'A:G')
    df['기준년월'] = '{}-{}'.format(yy, mm)
    
    ignore_list = ['아시아주','미주','구주','대양주','아프리카주','기타대륙',
                  '교포소계']
    condition = (df['국적'].isin(ignore_list))
    df_country = df[~condition].reset_index(drop = True)
    
    continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['대양주']*3 +\
                ['아프리카']*2 + ['기타']*2
    df_country['대륙'] = continents
    
    df_country['관광객 비율(%)'] = round(df_country['관광']/df_country['계']*100,
                                    ndigits = 2)
    
    df_country['전체비율(%)'] = round(df_country['관광']/sum(df_country['관광'])*100,
                                  ndigits = 2)
    
    return df_country
```

```python
df = pd.DataFrame()

for yy in range(2010,2021):
    for mm in range(1,13):
        try:
            mm_str = str(mm).zfill(2)
            yymm = '{}{}'.format(yy, mm_str)
            temp = create_kto_data(str(yy), mm_str)
            df = df.append(temp, ignore_index = True)
        except:
            pass
```
