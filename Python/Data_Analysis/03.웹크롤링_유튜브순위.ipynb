```python
# url = 'http://youtube-rank.com/board/bbs/board.php?bo_table=youtube'
# ['음악/댄스','가수명','구독자수','View수','동영상수']
# ['장르','채널명','구독자수','View수','동영상수']
#youtube_rank_class.xlsx

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup

import pandas as pd
from pandas import Series, DataFrame
```

```python
ser = Service('../chromedriver/chromedriver.exe')

driver = webdriver.Chrome(service = ser)

url = 'http://youtube-rank.com/board/bbs/board.php?bo_table=youtube'

driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html)
soup.head()
```

```python
char = soup.select('tbody > tr.aos-init')
char[:10]
```

```python
char[0].select('.subscriber_cnt')[0].text # 구독자수
```

```python
char[0].select('.video_cnt')[0].text # 영상수
```

```python
list = []
rank = 1
for you in char:
    genre = you.select('.category')[0].text.strip()[1:-1] # 장르
    name = you.select('a')[1].text.strip() # 채널명
    follower = you.select('.subscriber_cnt')[0].text # 구독자수
    view = you.select('.view_cnt')[0].text # 조회수
    video = you.select('.video_cnt')[0].text # 영상수
    
    print('{0}위'.format(rank),genre,name,follower,view,video,sep = ' | ')
    
    list.append([rank,genre,name,follower,view,video])
    rank += 1

list
```

```python
youtube_rank = pd.DataFrame(list,
                    columns = ['순위','카테고리','채널명','구독자수','조회수','동영상 수'])
youtube_rank
```

```python
youtube_rank.to_excel('youtube_rank_class.xlsx', index = False)
```
