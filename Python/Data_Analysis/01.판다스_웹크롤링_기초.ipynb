```python
sample_codemaster = pd.read_excel('./files/sample_codemaster.xlsx',
                                 header = 0,
                                 usecols = 'A:B',
                                 skipfooter = 0)
sample_codemaster
```

```python
sample_codemaster.head()
```

```python
# jupyter notebook 메소드 설명 : shift + tab
sample_1_code = pd.merge(left = sample_1,
                         right = sample_codemaster,
                         how = 'left',
                         left_on = '국적코드',
                         right_on = '국적코드'
                         )
sample_1_code
```

## merge how
1. inner : 교집합. 나머지 버림
2. outer : 합집합
3. left : 왼쪽으로
4. right : 오른쪽으로

```python
sample_1_code
```

```python
sample_2_code = pd.read_excel("./files/sample_2.xlsx",
                             header = 1,
                             usecols='A:C',
                             skipfooter=2
                             )

sample_2_code['기준년월'] = '2019-11'

sample_2_code = pd.merge(left = sample_2_code,
                        right = sample_codemaster,
                        how = 'left')
sample_2_code
```

```python
sample_2_code['test'] = 'A'
sample_1_append = sample_1_code.append(sample_2_code,
                                      ignore_index = True)

sample_1_append

# sample_concat = pd.concat([sample_1_code, sample_2_code],
#                          ignore_index = True,
#                          axis = 0)

# sample_concat
```

```python
sample_1_append.to_excel('./files/sample_class.xlsx',
                        index = False,
                        na_rep = 'NaN',
                        sheet_name = 'KNB_day1')
sample = sample_1_append
```

```python
with pd.ExcelWriter('./files/multiple_sheet.xlsx') as writer:
    sample.to_excel(writer, sheet_name = 'my_sheet1')
    sample.to_excel(writer, sheet_name = 'my_sheet2',
                   index = False, na_rep = 'NaN')
```

```python
## 피벗 테이블(pivot table)
sample_pivot = sample.pivot_table(values = '입국객수',
                                 index = '국적명',
                                 columns = '기준년월',
                                 aggfunc = 'mean')
sample_pivot
```

```python
#sample
# type(sample_pivot) << dataframe

import matplotlib.pyplot as plt

sample_pivot.plot(kind = 'bar',
                 figsize = (10,8),
                 rot = 0)
plt.xlabel("")
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```
